## Polusi Udara dan Manusia

Polusi seringkali dikaitkan pada perubahan iklim (i.e. permasalahan di masa depan). Opini publik
mengenai perubahan iklim sendiri beragam, di mana tidak sedikit masyarakat yang tidak mengakui
adanya perubahan iklim, atau perubahan iklim yang disebabkan oleh aktivitas manusia.

Meskipun begitu, polusi udara berkontribusi terhadap berbagai masalah pernapasan seperti asma,
penyakit paru obstruktif kronis, pneunomia, hingga tuberkulosis. Meskipun hubungan antara polusi
udara dan penyakit pernapasan tidaklah sederhana, studi epidemiologi baru-baru ini menemukan asosiasi
tingkat polusi udara yang disebabkan oleh lalu lintas terhadap peningkatan kasus asma dan penyakit
pernapasan lainnya.

Kasus Ella, dan masih banyak lainnya, sudah selayaknya menjadi pengingat bagi umat manusia bahwa,
terlepas dari segala "kontroversi" mengenai aktivisme polusi udara dan perubahan iklim, aktivitas
manusia yang mempengaruhi tingkat kualitas udara memiliki dampak langsung terhadap keberlangsungan
umat manusia. Laporan State of Global Air memperkirakan bahwa polusi udara berkontribusi terhadap
sekitar 6.67 juta kematian secara global pada tahun 2019. Polusi udara juga diperkirakan menurunan
rerata ekspektansi hidup manusia secara global hingga 1 tahun 8 bulan, kematian sekitar setengah
juta bayi secara global setiap tahunnya, dan meningkatkan risiko kesehatan bagi masyarakat rentan
seperti populasi berumur (i.e. lansia).

Bumi, planet yang diteorikan terbentuk dari ledakan kosmos, tidak perlu "diselamatkan"; ia telah

## Apa dan Mengapa?

Langkah pertama dalam menangani polusi udara adalah memperoleh pengetahuan mengenai kualitas udara
di lingkungan sekitar. Kami melakukan analisis deret waktu pada data kualitas udara yang dikumpulkan
oleh Air Quality Monitoring Station (AQMS) di Kota Athena, Yunani. Melalui analisis ini, diharapkan
terdapat insight mengenai bagaimana masing-masing metrik kualitas udara berkaitan satu sama lain,
dan bagaimana tren kualitas udara dalam selang waktu pengamatan.

Menggunakan algoritma pembelajaran, kami juga berharap untuk menemukan suatu model yang dapat
memprediksi kadar polutan PM2.5 dalam suatu waktu berdasarkan konsentrasi gas berbahaya yang
dideteksi. Particulate Matter (PM2.5) adalah partikel udara yang berukuran lebih kecil dari atau
sama dengan 2.5 µm. Secara tradisional, konsentrasi PM2.5 diukur menggunakan metode penyinaran
sinar Beta (Beta Attenuation Monitoring) dengan satuan mikrogram per meter kubik (µm/m3).

Dari semua polutan dalam udara, PM2.5 polusi menjadi ancaman kesehatan terbesar. Karena ukurannya
yang kecil, PM2.5 dapat bertahan di udara untuk waktu yang lama dan dapat masuk ke dalam tubuh
hingga ke dalam aliran darah saat dihirup.

### Date Parsing dan Sampling

Ketika bekerja dengan data time series, praktik yang umum adalah menjadikan kolom waktu sebagai
indeks, sehingga kolom date perlu di-parse dan dijadikan sebagai indeks. Selain itu, dataset berisi

## Data Visualization

### Spatial Distribution

Karena data kualitas udara diambil oleh beberapa AQMS yang tersebar di Kota Athena, mempelajari
distribusi spasial dari AQMS menjadi penting untuk memastikan bahwa data sensor yang didapat
merepresentasikan Athena dan bukan satu bagian kecil dari Athena.

```python
stations = df.groupby('station_name')[['PM2.5', 'Longitude', 'Latitude']].mean()

geometry = [Point(xy) for xy in zip(stations['Longitude'], stations['Latitude'])]
gdf = gpd.GeoDataFrame(stations, geometry=geometry)
gdf.set_crs(epsg=4326, inplace=True)
gdf = gdf.to_crs(epsg=3857)

plt.figure(figsize=(18, 16))
sns.scatterplot(x=gdf.geometry.x, y=gdf.geometry.y, hue=gdf.index, palette='tab20', s=100)
ctx.add_basemap(plt.gca(), crs=gdf.crs.to_string(), source=ctx.providers.CartoDB.PositronNoLabels)
plt.title('Spatial Distribution of Stations')
plt.legend(title='Station Name', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

AQMS cenderung terpusat pada pusat kota Athena yang padat penduduk, namun AQMS juga terdapat
pada daerah-daerah lain sehingga hasil pembacaan sensor AQMS dapat dikatakan representatif.

### Correlation Matrix

Mengetahui bagaimana korelasi tiap fitur berpengaruh terhadap fitur lainnya dapat memberikan
gambaran mengenai karakteristik dari sebuah metrik. Jika terdapat korelasi kuat antara fitur
terhadap target, maka dapat dipastikan fitur memiliki peranan penting dalam prediksi.

```python
correlation_matrix = df.drop(['id', 'code', 'station_name'], axis=1).corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

Terdapat beberapa insight yang diperoleh dari matriks korelasi:

- Temperatur berbanding terbalik dengan kelembaban (tentu saja!)
- Konsentrasi ground level Ozone dipengaruhi oleh temperatur dan kelembaban. Semakin panas suhu,
  semakin kering lingkungan, semakin tinggi konsentrasi ground level Ozone
- Ground level Ozone sendiri berpengaruh, kendati tidak secara signifikan, terhadap konsentrasi
  polutan PM10 dan PM2.5
- Konsentrasi O3 serta NO2 memiliki korelasi linear terhadap PM10 dan PM2.5
- Konsentrasi PM2.5 sangat dipengaruhi PM10. Hal ini masuk akal, karena PM2.5 secara teknis juga
  merupakan bagian dari PM10.
